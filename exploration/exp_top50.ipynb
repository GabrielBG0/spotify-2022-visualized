```python
import pandas as pd
import warnings
import seaborn as sns
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
pd.set_option('display.precision', 2)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
pd.set_option('display.max_columns', None)
sns.set_style("darkgrid")
sns.set_palette("colorblind")
sns.set_context("poster", font_scale=1, rc={
                "grid.linewidth": 5, "axes.linewidth": 5})
```

```python
df = pd.read_csv('top_songs_2022.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.corr()
```

```python
plt.figure(figsize=(18, 18))
heatmap = sns.heatmap(df.corr(), vmin=-1, vmax=1,
                      annot=True, fmt='.2f', annot_kws={"size": 13}, cmap='icefire')
heatmap.set_title('Corelations', fontdict={'fontsize': 20}, pad=20)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
ax = sns.histplot(data=df, x="loudness", kde=True)
ax.legend(['Densidade de Probabilidade (KDE)'], prop={'size': 20})
```

```python
plt.figure(figsize=(10, 8))
ax = sns.histplot(data=df, x="danceability", kde=True, bins=10)
ax.legend(['Densidade de Probabilidade (KDE)'], prop={'size': 20})
```

```python
plt.figure(figsize=(10, 8))
ax = sns.histplot(data=df, x="acousticness", kde=True)
ax.legend(['Densidade de Probabilidade (KDE)'], prop={'size': 20})
```

```python
plt.figure(figsize=(10, 8))
ax = sns.histplot(data=df, x="liveness", kde=True)
ax.legend(['Densidade de Probabilidade (KDE)'], prop={'size': 20})
```

```python
groupBy_key = df.groupby('key')
q_key = groupBy_key.size()
q_key
```

```python
plt.figure(figsize=(20, 5))
ax = sns.barplot(x=q_key.index, y=q_key, color='#7099E6')
ax.set_title('Songs by Key')
ax.set_ylabel('N of Songs')
ax.set_xticklabels(['C', 'C#/Db', 'D', 'D#/Eb',
                   'E', 'F', 'F#/Gb', 'G', 'G#/Ab', 'A', 'A#/Bb', 'B'])
plt.show()
```

```python
groupBy_mode = df.groupby('mode')
q_mode = groupBy_mode.size()
q_mode
```

```python
plt.figure(figsize=(10, 5))
ax = sns.barplot(x=q_mode.index, y=q_mode, color='#7099E6')
ax.set_title('Valores por Bairros')
ax.set_ylabel('Valor')
ax.set_xticklabels(['Minor', 'Major'])
plt.show()
```
